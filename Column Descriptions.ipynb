# COSMOS2025 ML Classification Catalog - Column Descriptions

## **Basic Identifiers**
| Column | Format | Description |
|--------|--------|-------------|
| `ra` | float64 | Right Ascension (J2000) in decimal degrees |
| `dec` | float64 | Declination (J2000) in decimal degrees |

## **Observed Magnitudes** (AB system)
| Column | Format | Description |
|--------|--------|-------------|
| `hst_f814w`    | float32 | HST/ACS F814W (I-band) magnitude |
| `uvista_y`     | float32 | VISTA Y-band magnitude |
| `uvista_j`     | float32 | VISTA J-band magnitude |
| `uvista_h`     | float32 | VISTA H-band magnitude |
| `uvista_ks`    | float32 | VISTA K<sub>s</sub>-band magnitude |
| `irac_ch1`     | float32 | Spitzer/IRAC Channel 1 (3.6μm) magnitude |
| `irac_ch2`     | float32 | Spitzer/IRAC Channel 2 (4.5μm) magnitude |
| `nircam_f115w` | float32 | NIRCAM F115w band magnitude |
| `nircam_f150w` | float32 | NIRCAM F150w band magnitude |
| `nircam_f227w` | float32 | NIRCAM F277w band magnitude |
| `nircam_f444w` | float32 | NIRCAM F444w band magnitude |

## **Derived Properties**
| Column | Format | Description |
|--------|--------|-------------|
| `zpdf_med`    | float32 | Median photometric redshift from LePhare |
| `mass_med`    | float32 | Median stellar mass (log(M<sub>⋆</sub>/M<sub>⊙</sub>)) from LePhare |
| `ssfr_med`    | float32 | Median specific star formation rate from LePhare |

## **Classification Flags**
| Column | Format | Description |
|--------|--------|-------------|
| `ML-class`        | bool | Machine learning classification (True=Quiescent, False=Non-Quiescent) |
| `SED-class`       | bool | SED-fitting classification (True=Quiescent, False=Non-Quiescent) |
| `ML-class_prob`   | bool | Probability of Machine learning classification |

## **Notes**
1. All magnitudes are in the AB system
2. ML classifications come from the CatBoost model
